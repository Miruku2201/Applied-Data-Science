```python
nlp = spacy.load("en_core_web_lg")
skill_pattern_path = "../jz_skill_patterns.jsonl"
```

```python
ruler = nlp.create_pipe('entity_ruler')
ruler.from_disk(skill_pattern_path)
nlp.add_pipe(ruler, before='ner')
nlp.pipe_names
```

```python
def get_skills(text):
    doc = nlp(text)
    myset = []
    subset = []
    for ent in doc.ents:
        if ent.label_ == "SKILL":
            subset.append(ent.text)
    myset.append(subset)
    return subset


def unique_skills(x):
    return list(set(x))
```

```python
clean = []
review = re.sub(
    '(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?"',
    " ",
    text,
)
review = review.lower()
review = review.split()
lm = WordNetLemmatizer()
review = [
    lm.lemmatize(word)
    for word in review
    if not word in set(stopwords.words("english"))
]
review = " ".join(review)
clean.append(review)
```
